## Estructuras de control

`Python` trabaja con indentación en vez de parentesis de llave para encapsular lo que está dentro de una estructura de control (ciclo, función, etc). La identación es un tabulador o 4 espacios sucesivos a la linea siguiente (enter) de la estructura de control. Es decir, algo que sería de la siguiente forma:


```python
if condicion:
    acción
```
**¡Notar los 4 espacios debajo del if!**

````python
nota = float(input("Hey, qué nota obtuviste en el taller de Intro a Py?: ")) # float numeros decimales
if nota >= 5.5:
    print("Eximido. Ud aprobó el curso")
elif nota >=4:
    print("Vaya estudair, debe dar examen")
else: 
    print("Has reprobado")
````

Los ciclos en `python` se caracterizan por tener una sintaxis un poco distinta, aunque intuitiva cuando uno se acostumbra. Tradicionalmente, uno define un índice que recorre un conjunto de valores `for(i=0, i<n, i++)`. Python recibe 
- Una lista de valores sobre los que itera
- El nombre de una variable que irá recorriendo esa lista
- Las instrucciones que se deben ejecutar

````python
print("Meses:", meses)

for i in meses:
    print("Mes:", i)
````

## Iteradores clásicos

Los iteradores son objetos cuyos valores aún no son generados, o sea contienen la información necesaria para producirlos cuando se utilice una estructura `for` o se le aplique la función `list` en el futuro. Esto es útil en términos de memoria. Por ejemplo, cuando cargamos 600,000 imágenes y aplicamos un filtro de colores, es conveniente cargar la foto, aplicar filtro, guardarla con el nuevo filtro a la vez y no todas al mismo tiempo.
Una vez que iteremos sobre el iterador, debemos definirlo nuevamente.

Algunas funciones que retornan iteradores son:

* `range(i, f, s)`: Crea una secuencia de datos desde `i` hasta `f` (este último no inclusivo con un salto `s`).
* `enumerate(lista)`: Retorna un iterador compuesta por tuplas que contienen el par `posición`, `elemento`.
* `zip(lista1, lista2)`: Retorna un iterador compuesto por tuplas que contienen elementos en la misma posición.

**Ejemplo range**


````python
# Secuencia de valores desde el 2 hasta el 9 de salto 2.
secuencia = range(2, 100, 2) 
print(secuencia)
````

El resultado es un iterador: para ver los valores, debemos aplicar la función `list` o un ciclo `for`. 

````python
print(list(secuencia))
````

````python
secuencia = range(2, 15, 2)
for i in secuencia:
    print("Valor:", i)
````

Ejemplo: 
Cómo construir un triángulo rectángulo como el de más abajo, tal que la altura me la entregue el usuario *(hint: puede hacer operaciones dentro de un print()).*

```
*
**
***
****
*****
```

````python
altura = int(input("Aqui va la altura del triángulo: "))
for i in range(altura):
   print("*"*(i+1))
````

````python
#Uso de \n y end="" en print

print("Esto es una linea \n pero acá hay otra linea")

altura = int(input("Aqui va la altura del triángulo: "))
for i in range(altura):
   print("*"*(i+1), end='@')
````

````python
altura = int(input("Introduce la altura del triángulo (entero positivo): "))
for i in range(altura):
    for j in range(i+1):
        print("*", end='')
    print('')
````

## Creación de funciones

Las funciones en python se trabajan a través del comando `def`

```python
def nombre_funcion(par1, par2,...):
    accion
    ...
    return valor
```

````python
def fahrenheit_a_celcius(temperatura):
  calculo = (temperatura - 32) / 1.8
  return calculo
````

````python
fahrenheit_a_celcius(3)
````

````python
#Escribir una función que reciba un número entero positivo y devuelva su factorial.
def factorial(n):
    """Función que calcula el factorial de un número entero positivo.
    Parámetros
    n: Es un entero positivo.
    Devuelve el factorial de n.
    """
    tmp = 1
    for i in range(n):
        tmp *= i+1
    return tmp

print(factorial(4))
print(factorial(20))
````
